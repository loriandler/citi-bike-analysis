# CitiBike Data Analysis

-	Combinine data from 2018-2023 in March for CitiBike data in NYC.
  
---

```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import swifter
```

```python
# File to Load 
file_path_2014 = Path("201403-citibike-tripdata.csv")
file_path_2015 = Path("201503-citibike-tripdata2.csv")
file_path_2016 = Path("201603-citibike-tripdata.csv")
file_path_2017 = Path("201703-citibike-tripdata.csv")
file_path_2018 = Path("201803-citibike-tripdata.csv")
file_path_2019 = Path("201903-citibike-tripdata.csv")
file_path_2020 = Path("202003-citibike-tripdata.csv")



# Read School and Student Data File and store into Pandas DataFrames
data2014_data = pd.read_csv(file_path_2014)
data2015_data = pd.read_csv(file_path_2015)
data2016_data = pd.read_csv(file_path_2016)
data2017_data = pd.read_csv(file_path_2017)
data2018_data = pd.read_csv(file_path_2018)
data2019_data = pd.read_csv(file_path_2019)
data2020_data = pd.read_csv(file_path_2020)


combined_tripdata_df = pd.concat([data2014_data, data2015_data, data2016_data, data2017_data, data2018_data, data2019_data, data2020_data])
print(combined_tripdata_df)

# combined_tripdata_df.to_csv('2014to2020.csv', index=False)
```

```python
# Function to parse date strings with different formats
def parse_dates(date_string):
    try:
        # Attempt to parse the date with the first format
        return pd.to_datetime(date_string, format='%m/%d/%Y %H:%M:%S')
    except ValueError:
        try:
            # If the first format fails, attempt to parse with the second format
            return pd.to_datetime(date_string, format='%Y-%m-%d %H:%M:%S')
        except ValueError:
            # Handle any other formats here if necessary
            return pd.NaT  # Return Not a Time for unparseable strings

# Apply the function to 'starttime' and 'stoptime' columns using swifter
combined_tripdata_df['starttime'] = combined_tripdata_df['starttime'].swifter.apply(parse_dates)
combined_tripdata_df['stoptime'] = combined_tripdata_df['stoptime'].swifter.apply(parse_dates)

# Drop rows where either 'starttime' or 'stoptime' is NaT (Not a Time)
combined_tripdata_df = combined_tripdata_df.dropna(subset=['starttime', 'stoptime'])
```

```python
# Trips over time
# Convert 'starttime' column to datetime format
combined_tripdata_df['starttime'] = pd.to_datetime(combined_tripdata_df['starttime'])

# Extract year and month from 'starttime'
combined_tripdata_df['year'] = combined_tripdata_df['starttime'].dt.year
combined_tripdata_df['month'] = combined_tripdata_df['starttime'].dt.month

# Group by year and month, and count the number of trips
trips_over_time = combined_tripdata_df.groupby(['year', 'month']).size().reset_index(name='num_trips')

# Plotting the trends
plt.figure(figsize=(10, 6))
plt.plot(trips_over_time['num_trips'])
plt.xlabel('Year-Month')
plt.ylabel('Number of Trips')
plt.title('Trends in Number of Trips Over Time')
plt.xticks(range(len(trips_over_time)), trips_over_time['year'].astype(str) + '-' + trips_over_time['month'].astype(str), rotation=45)
plt.tight_layout()
plt.show()
```
